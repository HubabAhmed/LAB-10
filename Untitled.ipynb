```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
image = cv2.imread('Capture.png', cv2.IMREAD_GRAYSCALE)
horizontal_mask = np.array([[-1, -1, -1],
                             [ 2,  2,  2],
                             [-1, -1, -1]])
vertical_mask = np.array([[-1,  2, -1],
                          [-1,  2, -1],
                          [-1,  2, -1]])
diagonal_mask1 = np.array([[ 2, -1, -1],
                           [-1,  2, -1],
                           [-1, -1,  2]])
diagonal_mask2 = np.array([[-1, -1,  2],
                           [-1,  2, -1],
                           [ 2, -1, -1]])
horizontal_edges = cv2.filter2D(image, -1, horizontal_mask)
vertical_edges = cv2.filter2D(image, -1, vertical_mask)
diagonal_edges1 = cv2.filter2D(image, -1, diagonal_mask1)
diagonal_edges2 = cv2.filter2D(image, -1, diagonal_mask2)
titles = ['Original Image', 'Horizontal Edges', 'Vertical Edges', 
          'Diagonal Edges 1', 'Diagonal Edges 2']
images = [image, horizontal_edges, vertical_edges, diagonal_edges1, diagonal_edges2]
plt.figure(figsize=(12, 8))
for i in range(len(images)):
    plt.subplot(2, 3, i+1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread('Tiger.jpg', cv2.IMREAD_GRAYSCALE)
sobel_horizontal = cv2.Sobel(img, ddepth=cv2.CV_64F, dx=1, dy=0, ksize=3)  # Horizontal edges
sobel_vertical = cv2.Sobel(img, ddepth=cv2.CV_64F, dx=0, dy=1, ksize=3)    # Vertical edges
sobel_combined = cv2.magnitude(sobel_horizontal, sobel_vertical)
sobel_horizontal = cv2.convertScaleAbs(sobel_horizontal)
sobel_vertical = cv2.convertScaleAbs(sobel_vertical)
sobel_combined = cv2.convertScaleAbs(sobel_combined)
titles = ['Original Image', 'Horizontal Edges', 'Vertical Edges', 'Combined Edges']
images = [img, sobel_horizontal, sobel_vertical, sobel_combined]
plt.figure(figsize=(10, 8))
for i, (image, title) in enumerate(zip(images, titles), start=1):
    plt.subplot(2, 2, i)
    plt.imshow(image, cmap='gray')
    plt.title(title)
    plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
image = cv2.imread('Tiger.jpg', cv2.IMREAD_GRAYSCALE)
grad_x = cv2.Sobel(src=image, ddepth=cv2.CV_64F, dx=1, dy=0, ksize=3)  # Gradient in X-direction (Horizontal edges)
grad_y = cv2.Sobel(src=image, ddepth=cv2.CV_64F, dx=0, dy=1, ksize=3)  # Gradient in Y-direction (Vertical edges)
combined_grad = cv2.magnitude(grad_x, grad_y)
horizontal_edges = cv2.convertScaleAbs(grad_x)
vertical_edges = cv2.convertScaleAbs(grad_y)
combined_edges = cv2.convertScaleAbs(combined_grad)
result_titles = ['Original Image', 'Horizontal Gradient', 'Vertical Gradient', 'Combined Gradients']
result_images = [image, horizontal_edges, vertical_edges, combined_edges]
plt.figure(figsize=(12, 8))
for idx, (img, title) in enumerate(zip(result_images, result_titles), 1):
    plt.subplot(2, 2, idx)
    plt.imshow(img, cmap='gray')
    plt.title(title)
    plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
input_image = cv2.imread('coin.png', cv2.IMREAD_GRAYSCALE)
smoothed_image = cv2.GaussianBlur(input_image, (5, 5), 0)
detected_edges = cv2.Canny(smoothed_image, 50, 150)
_, thresholded_image = cv2.threshold(smoothed_image, 120, 255, cv2.THRESH_BINARY)
morph_kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (3, 3))
processed_image = cv2.morphologyEx(thresholded_image, cv2.MORPH_CLOSE, morph_kernel)
combined_output = cv2.bitwise_and(processed_image, processed_image, mask=detected_edges)
image_titles = [
    'Grayscale Input', 'Smoothed Image', 
    'Edge Detection', 
    'Thresholded Image', 
    'Morphological Closing', 
    'Final Combined Output'
]
image_list = [
    input_image, 
    smoothed_image, 
    detected_edges, 
    thresholded_image, 
    processed_image, 
    combined_output
]
plt.figure(figsize=(12, 10))
for i, (img, title) in enumerate(zip(image_list, image_titles), start=1):
    plt.subplot(2, 3, i)
    plt.imshow(img, cmap='gray')
    plt.title(title)
    plt.axis('off')
plt.tight_layout()
plt.show()
```
